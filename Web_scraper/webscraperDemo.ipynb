## Dynamic website loading and automatic scrolling

```python
website = 'https://hcmut.edu.vn/danh-sach-tin-tuc'
path = 'C:\\Users\\admin\\Downloads\\chromedriver_win32\\chromedriver.exe'
```

```python
# service = Service()
driver = webdriver.Chrome(executable_path =  path)
driver.get(website)
```

```python
### OPTIONAL MODULE (Use when dynamic website requires loading by scrolling)
import time
SCROLL_PAUSE_TIME = 1

# INFINITE SCROLL (CAREFUL BEFORE USE)
# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

## Extract information from raw HTML

Header: h3 class heading p

Content: p  class tag block-ellipsis

```python
# Extract header and content from raw HTML
headers = soup.find_all('h3', class_ = 'heading')
headerContent = []
for header in headers:
    headerContent.append(header.find('p').text.lower())
print(headerContent)    

summaries = soup.find_all('p', class_ = 'tag block-ellipsis')
summaryContent = []
for summary in summaries:
    summaryContent.append(summary.text.lower())
print(summaryContent)

print(len(headerContent))
```

```python
# Print out website content
websiteContent = []
for i in range(len(summaries)):
    # print(f"Header: {headerContent[i]}")
    # print(f"Content: {summaryContent[i]}")
    header = headerContent[i]
    content = summaryContent[i]
    websiteContent.append([header, content])
```

## Save data to local files

```python
#Save into csv file
df = pd.DataFrame(websiteContent, columns=['Title', 'Summary'])
df.to_csv('tintuc.csv')
```

```python
#Save into separate txt file
#Remember to create a folder for txt files
for i in range(len(summaries)):
    with open(f'tintucs/tintuc{i}.txt', 'w') as f:
        f.write(f"{headerContent[i]}\n")
        f.write(f"{summaryContent[i]}")
    print(f'File saved {i}')
```
